```python
import pandas as pd
import numpy as np
import re

def clean_price(x):
    if pd.isna(x):
        return np.nan
    
    x = x.lower().replace('â‚¹', '').strip()
    
    # Extract numeric part
    number = re.findall(r'\d+\.?\d*', x)
    
    if not number:
        return np.nan
    
    number = float(number[0])
    
    if 'cr' in x:
        return number * 10000000
    
    elif 'l' in x or 'lac' in x or 'acs' in x:
        return number * 100000
    
    elif 'k' in x:
        return number * 1000
    
    else:
        return number

df['Price'] = df['Price'].apply(clean_price)
df['Price'] = df['Price'].astype(float)
```

```python
# Data
count_data = df["Residence_Type"].value_counts()
avg_sqft = df.groupby("Residence_Type")["Price_per_SQFT"].mean()

# Reindex to match same order
avg_sqft = avg_sqft.reindex(count_data.index)

fig, ax1 = plt.subplots(figsize=(8,5))

# Bar chart (Count)
bars = ax1.bar(count_data.index, count_data, color='skyblue')
ax1.set_ylabel("Property Count", color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Add count labels on bars
for bar in bars:
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2,
             height,
             f'{int(height)}',
             ha='center', va='bottom', fontsize=9)

# Line chart (Price per SQFT)
ax2 = ax1.twinx()
line = ax2.plot(avg_sqft.index, avg_sqft, color='red', marker='o', linewidth=2)
ax2.set_ylabel("Average Price per SQFT", color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Add values on line points
for i, value in enumerate(avg_sqft):
    ax2.text(i,
             value,
             f'{value:.0f}',
             color='red',
             ha='center', va='bottom', fontsize=9)

plt.title("Property Count vs Avg Price per SQFT")
plt.tight_layout()
plt.show()
```
